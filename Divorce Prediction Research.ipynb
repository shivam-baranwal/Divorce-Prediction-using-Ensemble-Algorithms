```python
#importing all the necessary libraries
from sklearn.datasets import make_classification
from sklearn.svm import SVC
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.ensemble import GradientBoostingClassifier
import pandas as pd
# import warnings filter
from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category=FutureWarning)
# import warnings filter
from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category=FutureWarning)
```

```python
#reading the dataset
df = pd.read_csv('Dataset.csv')
#splitting the dataset into X and Y variables
X,Y=df.iloc[0:,0:53],df.iloc[0:,-1]
#splitting the dataset into train and test in the ratio 0.67:0.33
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33,random_state=42)
#normalisation of the final dataset
norm=MinMaxScaler().fit(X_train)
X_train_norm=norm.transform(X_train)
X_test_norm=norm.transform(X_test)
```

```python
#Feature Selection using chi^2 and SelectKBest
bestatt=SelectKBest(score_func=chi2,k=6)
afit=bestatt.fit(X_train_norm,Y_train)
scrs=pd.DataFrame(afit.scores_)
cols=pd.DataFrame(X.columns)
chart=pd.concat([cols,scrs],axis=1)
chart.columns=['attribute','score']
print(chart.nlargest(6,'score'))
#Transforming the variables after feature selection
print("\nTraining Set:")
print("Before transf:",end='')
print(X_train_norm.shape)
X_train_new=bestatt.transform(X_train_norm)
print("After transf:",end='')
print(X_train_new.shape)
print("\nTest Set:")
print("Before transf:",end='')
print(X_test_norm.shape)
X_test_new=bestatt.transform(X_test_norm)
print("After transf:",end='')
print(X_test_new.shape)
print("\n")
print(X_train_new[0])
print("\nNew")
print(X_test_new[0])
```

```python
#classification Algorithm in the following Sequence:
#1. SVM using RBF and Linear Kernel
#2. Naive Bayes using Multinomial, Bernoulli and Gaussian functions
#3. Decision Tree classifier
#4. Random Forest Classifier
#5. Gradient Boosting
#6. ADA Boosting
```

```python
# 1.SVM Classifier using GridsearchCV
param={'C':[0.1,0.5,0.0001,0.005,0.0005,0.0001],'gamma':[0.0001,0.001,0.1,0.0005,0.005,0.05],'kernel':['linear','rbf','poly']}
svmclf=GridSearchCV(SVC(),param,refit=True,verbose=3,return_train_score=False)
svmclf.fit(X_train_new,Y_train)
```

```python
#creating dataframe
mdetails=pd.DataFrame(svmclf.cv_results_)
#sorting in decreasing order
mdetails=mdetails.sort_values('mean_test_score',ascending=False)
mdetails[['param_C','param_kernel','params','mean_test_score']]
```

```python
#Fitting the best model to the test set
svm_acc1=svmclf.score(X_test_new,Y_test)
svm_acc2=svmclf.score(X_train_new,Y_train)
print("SVM-Accuracy of on the test set:",end='')
print(svm_acc1)
print("SVM-Accuracy of on the train set:",end='')
print(svm_acc2)
print(classification_report(Y_test,svmclf.predict(X_test_new)))
```

```python
#end of SVM classifier
```

```python
# 2.1 Gaussian Naive Bayes
gnbclf=GaussianNB()
gnbclf.fit(X_train_new,Y_train)
#Fitting the best model to the test set
gnb_acc1=gnbclf.score(X_test_new,Y_test)
gnb_acc2=gnbclf.score(X_train_new,Y_train)
print("GNB-Accuracy of on the test set:",end='')
print(gnb_acc1)
print("GNB-Accuracy of on the train set:",end='')
print(gnb_acc2)
print(classification_report(Y_test,gnbclf.predict(X_test_new)))
```

```python
# 2.2 Bernoulli Naive Bayes
bnbclf=BernoulliNB()
bnbclf.fit(X_train_new,Y_train)
bnb_acc1=bnbclf.score(X_test_new,Y_test)
bnb_acc2=bnbclf.score(X_train_new,Y_train)
print("BNB-Accuracy of on the test set:",end='')
print(bnb_acc1)
print("BNB-Accuracy of on the train set:",end='')
print(bnb_acc2)
print(classification_report(Y_test,bnbclf.predict(X_test_new)))
```

```python
# 2.3 Multinomial Naive Bayes
mnbclf=MultinomialNB()
mnbclf.fit(X_train_new,Y_train)
mnb_acc1=mnbclf.score(X_test_new,Y_test)
mnb_acc2=mnbclf.score(X_train_new,Y_train)
print("MNB-Accuracy of on the test set:",end='')
print(mnb_acc1)
print("MNB-Accuracy of on the train set:",end='')
print(mnb_acc2)
print(classification_report(Y_test,mnbclf.predict(X_test_new)))
```

```python
#end of naive bayes classifier
```

```python
#3. Decision Tree Classifier using Grid Search
param={'max_depth':[1,2,3,4,5],'min_samples_split':[5,10,20,30,40,50],'min_samples_leaf':[1,2,3,4,5,10],'random_state':[1,15,42,0]}
dtclf=GridSearchCV(DecisionTreeClassifier(),param,refit=True,verbose=3,return_train_score=False)
#clf = DecisionTreeClassifier(max_depth=3,min_samples_split=10,min_samples_leaf=4)
dtclf.fit(X_train_new, Y_train)

dtclf_acc1 = dtclf.score(X_test_new, Y_test)
dtclf_acc2 = dtclf.score(X_train_new, Y_train)

print("score on test: "  + str(dtclf_acc1))
print("score on train: " + str(dtclf_acc2))
print(classification_report(Y_test,dtclf.predict(X_test_new)))
```

```python
#4. Random Forest Classifier Using GridSearch
param={'n_estimators':[1,10,100,1000,10000],'max_depth':[1,2,3,4,5],'random_state':[1,15,42,0]}
rf=GridSearchCV(RandomForestClassifier(),param,refit=True,verbose=3,return_train_score=False)
#rf = RandomForestClassifier(n_estimators=1000, max_depth=2,random_state=42)
rf.fit(X_train_new, Y_train)

rf_acc1 = rf.score(X_test_new, Y_test)
rf_acc2 = rf.score(X_train_new, Y_train)


print("score on test: " + str(rf_acc1))
print("score on train: "+ str(rf_acc2))
print(classification_report(Y_test,rf.predict(X_test_new)))
```

```python
#details of hyperparameters used in Random Forest Classifiers
mdetails=pd.DataFrame(rf.cv_results_)
mdetails=mdetails.sort_values('mean_test_score',ascending=False)
mdetails
```

```python
#5. Gradient Boosting Classifier
param={'n_estimators':[1,10,50,100,1000],'learning_rate':[0.1,1.0,2.0,3.0,5.0,10.0],'max_depth':[1,2,3,4,5,10],'random_state':[1,15,42]}
gbclf=GridSearchCV(GradientBoostingClassifier(),param,refit=True,verbose=3,return_train_score=False)
gbclf.fit(X_train_new,Y_train)
gbclf_acc1=gbclf.score(X_test_new, Y_test)
gbclf_acc2=gbclf.score(X_train_new,Y_train)
print(gbclf_acc1)
print(gbclf_acc2)
mdetails=pd.DataFrame(gbclf.cv_results_)
mdetails=mdetails.sort_values('mean_test_score',ascending=False)
mdetails
```

```python
#6. ADA Boosting Classifier
from sklearn.ensemble import AdaBoostClassifier
param={'n_estimators':[10,50,20,100,1000],'learning_rate':[1,0.1,0.001,0.0001]}
abclf=GridSearchCV(AdaBoostClassifier(base_estimator=GaussianNB(),random_state=42),param,refit=True,verbose=3,return_train_score=False)
abclf.fit(X_train_new,Y_train)
mdetails=pd.DataFrame(abclf.cv_results_)
mdetails=mdetails.sort_values('mean_test_score',ascending=False)
mdetails[['params','mean_test_score']]
abclf_acc1 = abclf.score(X_test_new, Y_test)
abclf_acc2 = abclf.score(X_train_new, Y_train)
print("score on test: " + str(abclf_acc1))
print("score on train: "+ str(abclf_acc2))
print(classification_report(Y_test,abclf.predict(X_test_new)))
```
